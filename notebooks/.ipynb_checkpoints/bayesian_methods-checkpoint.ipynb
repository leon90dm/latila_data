```python
def get_avg(post_p):
    s = 0
    for i,j in enumerate(post_p):
        s += (i+1)*j
    return s
```

N vs 后验均值的图像

```python
postp_avg = [get_avg(get_postp(x)) for x in np.r_[500:5000:500]]

plt.plot([x for x in np.r_[500:5000:500]], postp_avg)
plt.scatter([x for x in np.r_[500:5000:500]], postp_avg)
plt.xlabel('prior $N$')
plt.ylabel('post $E[N]$')
```

除了看到No.60的列车，还看到了 No.30 、No.90 的。那么：

```python
N = 1000
def update_for_observed_data(N, data_list):
    def get_postp2(N, data, priors = None):
        priors = np.full(N, 1/N) if priors is None else priors
        pdata = np.asanyarray([0 if x < data else 1/x for x in range(N)])  
        likelyhood = priors * pdata
        post_p = likelyhood/likelyhood.sum()
        return post_p
    post_p = get_postp2(N, data_list[0])
    if len(data_list) > 1:
        for data in data_list[1:]:
            post_p = get_postp2(N, data, post_p)
    return post_p
```

```python
plt.plot(update_for_observed_data(N, data_list=[60,30,90]))
```

```python
postp_avg = [get_avg(update_for_observed_data(x, data_list=[60,30,90])) for x in np.r_[500:5000:500]]

plt.plot([x for x in np.r_[500:5000:500]], postp_avg)
plt.scatter([x for x in np.r_[500:5000:500]], postp_avg)
plt.xlabel('prior $N$')
plt.ylabel('post $E[N]$')
```

增加规模定律

$$
\operatorname{PMF}(N=x) \propto\left(\frac{1}{x}\right)^{\alpha}\ ; \text(\alpha \approx 1)
$$

```python
N = 1000
def update_for_observed_data_pmf(N, data_list, alpha=1):
    def get_init_prior(N,alpha):
        p = np.asarray([np.power(1/(x+1), alpha) for x in range(N)])
        return p/p.sum()
    def get_postp2(N, data, priors = None):
        priors = get_init_prior(N, alpha) if priors is None else priors
        pdata = np.asarray([0 if x < data else 1/x for x in range(N)])  
        likelyhood = priors * pdata
        post_p = likelyhood/likelyhood.sum()
        return post_p
    post_p = get_postp2(N, data_list[0])
    if len(data_list) > 1:
        for data in data_list[1:]:
            post_p = get_postp2(N, data, post_p)
    return post_p
```

```python
ax = plt.subplot()
ax.plot(update_for_observed_data_pmf(N, data_list=[60]), label='PMF with $\\alpha=1.0$')
ax.plot(update_for_observed_data(N, data_list=[60]), label='Uniform')
plt.xlabel('$N$')
plt.legend()
```

```python
postp_avg_pmf = [get_avg(update_for_observed_data_pmf(x, data_list=[60,30,90])) for x in np.r_[500:5000:500]]
postp_avg = [get_avg(update_for_observed_data(x, data_list=[60,30,90])) for x in np.r_[500:5000:500]]
ax = plt.subplot()
ax.plot([x for x in np.r_[500:5000:500]], postp_avg, label='Uniform')
ax.scatter([x for x in np.r_[500:5000:500]], postp_avg)
ax.plot([x for x in np.r_[500:5000:500]], postp_avg_pmf, label='PMF with $\\alpha=1.0$')
ax.scatter([x for x in np.r_[500:5000:500]], postp_avg_pmf)
plt.xlabel('prior $N$')
plt.ylabel('post $E[N]$')
plt.legend()
```

# 置信区间

```python
def percentile(pmf, percentage): 
    p = percentage / 100.0
    total = 0
    for val, prob in pmf.items():
        total += prob
        if total >= p:
            return val
```

```python
N=1000
post_p = pd.Series(update_for_observed_data_pmf(N, data_list=[60,30,90]), index=np.r_[1:N+1:1])
interval  = percentile(post_p, 5), percentile(post_p, 95)
print(interval)
```

在前面的示例(看到了三个火车,且呈幂律分布的先验概率的火车头问题)中90%置信区间为(92,243)。如此大的范围其实确切的表明,(尽管平均值收敛了)我们仍然相当不确定究竟有多少火车头存在。

# 欧元问题

>2000年1月4日,星期五,《卫报》上刊载了一个统计相关的声明: 当以边缘转动比利时一欧元硬币250次后,得到的结果是正面140次反面110次。“这看起来很可疑”,伦敦经济学院的统计讲师巴里·布莱特说,“如果硬币是均匀的,得到这个结果的可能性低于7%。”

定义假设， $H_x$ 表示硬币正面朝上的概率为 x%, 则 $x \in [0,100]$

```python
data = 'H'*140+'T'*110
data
```

```python
priors = np.full(101, 1/100)
for d in data:
    likelyhood = np.arange(101)/100
    if 'T' == d:
        likelyhood = 1-likelyhood  # 在这个正面朝上概率的假设下，得到反面的概率为 1-x%
    post_p = priors * likelyhood
    post_p = post_p/post_p.sum()
    priors = post_p
```
