### 但这和贝叶斯有什么关系？ 什么是似然函数？ 如何建模？

这里我们举个例子，以目前已有的数据，假设给出一个showcase， 我猜测价格为20,000元的先验和后验分布是什么样的？ 

$$ 
P(price=x | guess=20,000)  
$$

所以说我这里的data是： 猜测价格是20,000元， 而我们想得到的就是showcase为各种价格的假设的概率

似然函数：

$$
P(showcase价格=x | 猜测价格=y )
$$

这个似然函数更奇怪, 当我的猜测价格为y时，展示品价格为x的概率。例如：但我猜1元的时候，展示品为100元的概率。
我的猜测价格 能 影响展示品的价格分布么？

有没有什么经验可以借鉴的呢？

貌似有： 我们可以假设数据集中有很多玩家猜测（bid）的数据是基于玩家的经验而来，即 展品价格与猜测价格之间的误差应该服从某种分布的。由此

我们看下给的数据里面误差的分布：

```python
t[['Difference 1', 'Difference 2']].plot.kde()
```

我们可以看到两个选手的猜测误差（showcase - bid）比较服从正态分布，但偏右，这是游戏规则导致的，选手尽量往低价猜

所以我们假设error的误差分布是一个 均值为0， 且方差为现有Difference 估计的方差

```python
pdf_error = stats.norm(0, np.std(t['Difference 1']))
```

然后我们可以编写似然函数了， 似然函数输出的给定猜测价格下是showcase 价的概率

```python
# 先验概率分布
prior = stats.gaussian_kde(t['Showcase 1'])
```

```python
def prior_fn(x):
    return prior.pdf(x)[0]
def likelyhood(guess, price):
    return pdf_error.pdf(price - guess)
def posterior(guess, price):
    prior = prior.pdf(guess)
    like = likelyhood(guess, price)
    return prior*like
```

```python
n = 101
price_xs = pd.DataFrame(np.linspace(0, 75000, n), columns=['price'])

price_xs['prior'] = price_xs.price.apply(lambda x: prior_fn(x))
price_xs['likelyhood'] = price_xs.price.apply(lambda x: likelyhood(20000, x))
price_xs['posterior'] = price_xs.prior * price_xs.likelyhood

price_xs.posterior = price_xs.posterior/price_xs.posterior.sum()

price_xs.prior = price_xs.prior/price_xs.prior.sum()

price_xs = price_xs.set_index('price')

price_xs[['prior','posterior']].plot()
```
